## Checking Relation between all variables (Bivariate Analysis)

```python
num=[col for col in car_data.columns if col not in categorical]
for i in num:
    plt.figure(figsize=(12,4))
    sns.distplot(car_data[i],bins=int(180/5), color = 'darkblue',hist_kws={'edgecolor':'black'},kde_kws={'linewidth': 1})
    plt.title("Distribution of %s"%i,fontsize=20)
    plt.xticks(fontsize=15)
    plt.show()
    print('\n')
```

### Checking Relation between all categorical variables and dependent variables (price)

```python
"""Creating a function for ordering the groups in a column as per their frequency"""
def sort_order(column):
    orders = (car_data.groupby([column]).mean().sort_values(by ='Price', ascending = False)).index
    return orders
```

```python
"""Looping over categorical variables to check the Price over different Groups"""
for i in categorical:
    if car_data[i].nunique() < 72:
        plt.figure(figsize=(12,5))
        sns.barplot(car_data[i],car_data['Price'], order = sort_order(i), palette='copper')
        plt.title("Bar Plot of %s" %i, fontsize = 10)
        plt.xticks(fontsize = 1)
        plt.xlabel("%s"%i,fontsize=0.5)
        plt.ylabel("Car Price")
        plt.xticks(fontsize = 10, rotation = 90)
        plt.show();
        print('\n')
```
